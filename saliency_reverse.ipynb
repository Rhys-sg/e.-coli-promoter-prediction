```python
# Grid-based saliency map for Urtecho_2023_strand sequences that are not 100% confident when predicting the correct label
# It shows what the model is confused on and which features are important for the model to make a decision.

import pandas as pd

Urtecho_2023_strand = pd.read_csv('Data/Urtecho_2023_strand.csv')
Urtecho_2023_strand['Promoter Sequence'] = Urtecho_2023_strand['promoter_sequence']
Urtecho_2023_strand['Reversed'] = Urtecho_2023_strand['forward'].apply(lambda x: 1 if x == False else 0)

saliency_module.plot_isForward_saliency_map_grid(
    num_samples=100,
    relative=False,
    model_filename='Models/isForward_Urtecho_CNN.keras',
    data=Urtecho_2023_strand
)
```
